```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale 
```

```python
# Normalizing the numerical data 
wine_normal = scale(wine.data)
wine_normal
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
pca.components_[0]
```

```python
# Cumulative variance 

var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0]
y = pca_values[:,1]
z = pca_values[:2:9]
plt.scatter(x,y,color=["red","blue"])
```

```python
pca = PCA(n_components=3)
```

```python
PR_comp = pca.fit_transform(wine_normal)
PR_comp
```

```python
Principal_components = pd.DataFrame(data=PR_comp,columns=['Principal Component 1','Principal Component 2','Principal Component 3'])
Principal_components.head()
```

```python
#K means clustering
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1,20):
    Kmeans_pca = KMeans(n_clusters = i,init='k-means++',random_state=42)
    Kmeans_pca.fit(PR_comp)
    wcss.append(Kmeans_pca.inertia_)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,20),wcss,marker='o',linestyle='--')
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.title('Kmeans clustering with PCA')
```

```python
# Now that we know there are 3 clusters.

Kmeans_new = KMeans(n_clusters=3,init='k-means++',random_state=42)
```

```python
a = Kmeans_new.fit_predict(PR_comp)
a
```

```python
# Visualize the cluster using scatterplot
plt.figure(figsize=(10,8))
sns.scatterplot(PR_comp[a==0,0],PR_comp[a==0,1],label='cluster 1')
sns.scatterplot(PR_comp[a==1,0],PR_comp[a==1,1],label='cluster 2')
sns.scatterplot(PR_comp[a==2,0],PR_comp[a==2,1],label='cluster 3')
# create the centroids.
sns.scatterplot(Kmeans_new.cluster_centers_[:,0],Kmeans_new.cluster_centers_[:,1],label='Centroids',color='black')
plt.title('Clusters by PCA Components')
```

```python
import scipy.cluster.hierarchy as sch  
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
z = linkage(PR_comp, method='complete', metric='euclidean')
```

```python
plt.figure(figsize=(25,5))
plt.title('Hierarchial Dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_font_size=8,leaf_rotation=0)
plt.show()
```

```python
z = linkage(PR_comp, method='ward', metric='euclidean',optimal_ordering=False)
plt.figure(figsize=(25,10))
plt.title('Hierarchial of the PCA Components')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_font_size=8,leaf_rotation=0)
plt.show()
```
